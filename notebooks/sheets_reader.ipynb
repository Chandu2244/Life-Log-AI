```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# Step 1: Define scope (permission areas)
scope = ["https://www.googleapis.com/auth/spreadsheets",
         "https://www.googleapis.com/auth/drive"]

# Step 2: Load JSON key file
creds = ServiceAccountCredentials.from_json_keyfile_name(
    "dailylifeinsights-25eb2f336dd0.json", scope
)

# Step 3: Authorize client
client = gspread.authorize(creds)

# Step 4: Open sheet by name
sheet = client.open("DailyJournal").worksheet("Journal")   # sheet1 = first tab

# Step 5: Read all data
rows = sheet.get_all_values()

print(rows)
```
